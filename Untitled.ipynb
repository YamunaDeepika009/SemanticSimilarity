```python
# 1) Install (run once)
# pip install -U sentence-transformers numpy

import time
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
# 2) Toy corpus (replace later with a bigger dataset)
with open("sample.txt", "r", encoding="utf-8") as f:
    docs = [line.strip() for line in f if line.strip()]
```

```python
# 3) Load a pretrained embedding model (good default)
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
# 4) Encode documents
doc_emb = model.encode(docs, normalize_embeddings=True)  # (N, D)
```

```python
def search(query: str, top_k: int = 3):
    # 5) Encode query
    q = model.encode([query], normalize_embeddings=True)[0]  # (D,)

    # 6) Brute-force cosine via dot product (because normalized)
    t0 = time.time()
    scores = doc_emb @ q  # (N,)
    print(scores)
    idx = np.argsort(-scores)[:top_k]
    latency_ms = (time.time() - t0) * 1000

    results = [(int(i), float(scores[i]), docs[i]) for i in idx]
    return latency_ms, results
```

```python
# 7) Test
lat, results = search("Where did zara find the lamp?", top_k=3)
print(f"Latency: {lat:.2f} ms")
for i, s, d in results:
    print(f"[{i}] score={s:.3f} | {d}")
```
